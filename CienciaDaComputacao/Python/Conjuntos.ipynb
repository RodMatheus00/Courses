# Conjuntos

### Características
- Não ordenadas
- Mutáveis
- Aceitam diferentes elementos
- Elementos únicos

### Construtores
- set()
- {}

```python
conj = set(["Python", "C", "C++", "Java", "R", "R", "C", "C"])

print(conj)
```

```python
conj2 = {"ABC", 32, 32, 7.2, False}

print(type(conj2))

print(conj2)
```

## Não Ordenadas

```python
for el in conj2:
    print(el)
    # outro comando
```

### Métodos

### Add
Adiciona um elemento ao conjunto

```python
# Conjuntos são estruturas sem duplicatas
conj2.add(44)
conj2.add(44)
conj2.add(44)

print(conj2)

for teste in conj2:
    print(teste)
```

### Remove 
Remove um elemento do conjunto

```python
# Remove in-place
conj2.remove("ABC")

print(conj2)

for teste2 in conj2:
    print(teste2)
```

### Intersection
Retorna os elementos em comum aos dois conjuntos

```python
print(conj)
print(conj2)

conj.intersection(conj2)
```

```python
conj2.add("Python")

conj.intersection(conj2)
```

### Union
Retorna um conjunto contendo todos os elementos de ambos conjuntos

```python
print(conj)
print(conj2)

conj.union(conj2)
```

## Difference
c1.difference(c2)

Retorna todos os elementos presentes somente no conjunto c1.

```python
print(conj)
print(conj2)

conj.difference(conj2)
```
