# Conjuntos

## Características

- Não ordenádas
- Mutáveis
- Aceitam diferentes elementos
- **Elementos únicos**

## Construtores

- `set()`
- `{}`

```python
conj = set(["Python", "C", "C++", "Java", "R", "R", "C", "C"])

print(conj)
```

```python
conj2 = {"ABC", 32, 32, 7.2, False}

print(type(conj2))

print(conj2)
```

```python
# Conjuntos:
{}

# Dicionários:
{key:value}
```

## Métodos

### Add

Adiciona um elemento ao conjunto

```python
# Adiciona in-place
conj2.add(44)
conj2.add(44)
conj2.add(44)
conj2.add(44)
conj2.add(44)

print(conj2)
```

### Remove

Remove um elemento do conjunto.

```python
# Remoção in-place
conj2.remove("ABC")

print(conj2)
```

### Intersection

Retorna os elementos em comum aos dois conjuntos.

```python
print(conj)
print(conj2)

conj.intersection(conj2)
```

```python
conj2.add("Python")

conj.intersection(conj2)
```

### Union

Retorna um conjuntos contendo todos os elementos de ambos conjuntos.

```python
print(conj)
print(conj2)

conj.union(conj2)
```

### Difference

> c1.difference(c2)

Retorna todos os elementos presentes somente no conjunto `c1`.

```python
print(conj)
print(conj2)

conj.difference(conj2)
```
