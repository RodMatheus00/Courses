# Tipo Básicos

## Números

- Números flutuantes : Possuem casas decimais
- Inteiros : Representa a parte inteira do número

```python
# Números flutuantes
print(1.0)
print(type(1.0))
print(6.0)
print(type(6.0))
print(12.4)
print(type(12.4))
print(654.21)
print(type(654.21))
```

```python
# Números inteiros
print(1)
print(type(1))
print(5)
print(type(5))
print(10)
print(type(10))
print(212)
print(type(212))
```

# Strings
Strings são textos: um ou mais caracteres

```python
print("Elefante")
print(type("Elefante"))
print("Cachorro")
print(type("Cachorro"))
print("Aprendendo Python")
```

```python
# Testando aspas simples e duplas
print("Aprendendo 'Python'")
print('Aprendendo "Python"')
```

```python
# Escape Chacacters
print("Aprendendo \"Python\"")
print(type("Aprendendo \"Python\""))

print("Nova \n linha")
print(type("Nova \n linha"))

print("Elemento \t elemento")
print(type("Elemento \t elemento"))
```

## Booleanos
São dados que possuem dois valores: True or False

```python
# Valores Booleano verdadeiro
print(True)
print(type(True))

# Valores Booleano falso
print(False)
print(type(False))
```

```python
print(bool(1))
print(bool(0))
```
