# Loops

Estas estruturas altera o fluxo do programa ao determinar que um determinado código seja executado várias vezes.

## Iteráveis e Iterações

- Iteração: cada execução completa de um comando sobre um iterável.

- Iteráveis: objetos nos quais pode haver iteração, ou seja, podemos executar algum comando sobre todos seus elementos.
    - Listas, tuplas, dicionáŕios, conjuntos, strings, ...

## Range

O objeto `range` é uma estrutura única de Python que representa uma sequência de números inteiros.

```python
print(range(20))
print(type(range(20)))

print(list(range(20)))

# range(inicio:fim:passos)

print(list(range(5, 15, 2)))
```

## Dois tipos de loops

### Loops Controlados por Condições:

Neste tipo de estrutura, os comandos são executados até uma dada condição tornar-se `False`. **Não sabemos quantas iterações haverão a priori**.

### Loops Controlados por Contagem:

Neste tipo de estrutura, os comandos são executados até um dado número ser alcançado. **Sabemos quantas iterações haverão a priori**.

## For Loop

Uma estrutura de loop controlada por contagem. Deste modo, temos uma variável de contagem para determinar quantas iterações haverão e, então, parar o loop.

```python
for _ in range(20):
    print("Python")
```

```python
var = [1, 2, 3, 4, 5, 6]

for x in range(6):
    #print(f"x = {x}")
    print(var[x])
```

```python
var = [1,2,3,4,5,6]

for x in var:
    print(x)
```

## While Loop

Uma estrutura de loop controlada por condição. Sendo assim. temos uma condição para determinar quantas iterações haverão e, então, quando esta condição tornar-se falsa, parar o loop.

```python
while condição:
    código
```

```python
var = 0

while var < 6:
    print(var)
    var += 1 # Atualização (iadd: in-place)
```

# Continue

Comando utilizado não para parar o loop, mas sim, para **passar para a próxima iteração**.

```python
for x in range(11):
    
    if x == 5:
        continue
    
    print(x)
```

# Break

Comando utilizado para **parar o loop**.

```python
for x in range(11):
    
    if x == 5:
        break
    
    print(x)
```

```python
while True:
    
    # Pediria a senha
    
    if senha == correta:
        break
```
