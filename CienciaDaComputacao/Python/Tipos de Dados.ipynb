# Tipos Básicos

## Números

- Números flutuantes: possuem casas decimais
- Inteiros: a parte inteira

```python
# Números flutuantes:

print(8.0)
print(type(8.0))

print(6.0)
print(type(6.0))

print(12.4)
print(type(12.4))

print(654.21)
print(type(654.21))
```

```python
# Números Inteiros:

print(6)
print(type(6))

print(15)
print(type(15))

print(8756)
print(type(8756))
```

## Strings

Strings são texto: um ou mais caracteres

```python
print("elefante")

print('cachorro')
```

```python
print("Aprendendo Python!")
```

```python
print("Aprendendo 'Python'")
```

```python
# Escape characters

print("Aprendendo \"Python\"")

print("Nova \n linha")

print("Elemento \t elemento")
```

```python
print(type("string"))
```

## Booleanos

São dados que possuem dois valores somente: True e False.

```python
# Valor booleano verdadeiro:
print(True)
print(type(True))

print(False)
print(type(False))
```

```python
print(bool(0))

print(bool(1))

print(bool(-587))

print(bool(12.86))
```

```python
print(bool(""))

print(bool("a"))
```
