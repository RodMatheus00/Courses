# Variáveis

Variáveis são nomes que estão ligados a objetos na memória e fazem referência a estes.

Operador `=`: designação de variáveis

Operador `+=`: iadd (adição *in-place*)

```python
# Variável inteira:
numero_pessoas = 16

print(numero_pessoas)
```

```python
# Variável string:
linguagem = "python"

print(linguagem)
```

```python
numero_pessoas / 10
```

```python
print(linguagem + "java")

linguagem += linguagem + "java"

print(linguagem)
```

```python
#numero_pessoas += 4

#numero_pessoas *= 10

#numero_pessoas -= 50

numero_pessoas /= 100

print(numero_pessoas)
```

```python
print(5 + 9)

# F-string:

print(f"Resultado: {5 + 9}")

print(f"Variável linguagem: {linguagem}")

print(f"Variável linguagem:\n{linguagem}\n")
```
