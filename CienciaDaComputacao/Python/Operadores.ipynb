# Operadores
Operadores são estruturas utilizadas na comparação/manuseio de objetos e variáveis. Retornam valores booleanos.

## Operadores de Comparação
Utilizados para comparar VALORES entre objetos

## Maior que: >
Retorna True quando o operando da esquerda é maior que o operando da direita

```python
print(5 > 4)

print(5 < 4)

print(5 > 5.0)
```

## Maior ou igual que: >=
Retorna True quando o operando da esquerda é maior ou igual ao operando da direita

```python
print(5 >= 8)

print(8 >= 5)

print(8 >= 8.0)
```

## Menor que: <
Retorna True quando o operando da esquerda é menor que o operando da direita

print(4 < 5)

print(5 < 4)

## Menor ou igual que: <= 
Retorna True quando o operando da esquerda é menor ou igual ao operando da direita

```python
print(4 <= 4)

print(4 <= 6)
```

## Igualdade: == 
Retorna True quando ambos operando são iguais

```python
print(4 == 6)

print(6 == 6)
```

## Diferença: != 
Retorna True quando ambos operadores são diferentes

```python
print(4 != 5)

print(5 != 5)
```

## Obs.:
Também funcionam para variáveis

```python
v1 = "Python"

print(v1 == "Python")
```

## Operadores de Identidade
Utilizados para comparar se dois objetos são os mesmos

## Is: is
Retorna True se os dois objetos de fato forem os mesmos, ou seja, locarizarem-se no mesmo local de memória

```python
v2 = [10, 20, 30]

v3 = [10, 20, 30]

print(v2 is v3)

print(id(v2))
print(id(v3))

v2[0] = 150

print(v2)
print(v3)
```

```python
v4 = v5 = [1, 2, 3, 4]

print(id(v4))
print(id(v5))

print(v4 is v5)

v4[0] = 150

print(v4)
print(v5)
```

## Is Not: is not
Retorna True se os dois objetos não forem os mesmos, ou seja, locarizarem-se em locais diferentes da memória

```python
print(v2 is not v3)

print(v4 is not v5)
```

## Operadores de Filiação
Utilizados para verificar se um objeto pertence a (está contido em) outro objeto

## In: in
- ob1 in obj2

Retorna True se obj1 estiver contido em obj2

```python
print(v4)

print(150 in v4)

print(25 in v4)
```

## Not In: not in
- obj not in obj2

Retorna True se Obj1 não estiver contido em obj2

```python
print(v4)

print(150 not in v4)

print(25 not in v4)
```

### Dicionários:

```python
d = {"Linguagem": "Python", "Aprendendo": True}

print(d)
```

```python
print("Linguagem" in d)

print("Python" in d.values())
```

## Operadores Lógicos
Utilizados para concatenar comparação simples, criando entaõ comparações compostas

## And: and
- exp1 and exp2

Retorna True se ambas expressões forem verdadeiras

```python
print(5 > 2 and 6 > 4)

print((10 <= 100) and (55 != 55) and (89 > 5))
```

## Or: or
- Exp1 and exp2

Retorna True se alguma das expressões forem verdadeiras

```python
print(5 > 4 or 6 > 4)

print((10 <= 100) or (55 != 55))

print((10 >= 100) or (55 != 55))
```

## Not: not
- not exp1

Nega logicamente a expressão, ou seja, altera o valor lógico para seu oposto

```python
print(not True)
print(not False)

print(not(5 > 2) or (6 < 4))
```
