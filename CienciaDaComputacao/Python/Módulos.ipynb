# Módulos em Python

Módulos são scripts como qualquer outro, porém nestes escrevemos definições de funções e objetos. Uma vez que estão em um script à parte, podemos reutilizar estas funções e objetos em diversos programas diferentes sem ter de reescrevê-los toda vez.

Estas definições que estão dentro do módulo podem ser reutilizadas através de sua **importação** para o programa em questão. Depois que importamos, podemos utilizar esta função/objeto/... normalmente, como se tivéssemos definido-as no programa atual.

Importação de módulos em Python é feita com a palavra-chave `import`.

# Importando Módulos

Usando apenas `import`, todas as funcionalidades do módulos tornam-se disponíveis ao usuário através do operador `.`

```python
import math
```

```python
dir(math)
```

```python
print(math.pi)

print(math.log(2))

print(math.sin(56))

print(math.pow(4, 10))
print(4 ** 10)

print(math.sqrt(25))
```

```python
import random
```

```python
dir(random)
```

```python
print(random.choice([1,2,3,4]))

print(random.choice(range(1, 5)))
```

```python
print(random.choices(range(1, 5), k=3))
```

# From `módulo` import `nome`

Forma alternativa de se importar algo de um módulo. Com esta forma, temos acesso direto ao objeto importado, sem necessidade do operador `.`

```python
from random import shuffle
```

```python
help(shuffle)
```

```python
lst = list(range(11))

print(lst)

shuffle(lst)

print(lst)
```

## Importando vários objetos de uma vez

Desta mesma forma, podemos importar diversos objetos de um mesmo módulo separando-os por vírgulas.

```python
from random import choice, shuffle
```

```python
lst = list(range(11))

print(lst)

shuffle(lst)

print(lst)

print(choice(lst))
```

# Import `módulo` as `nome_alternativo`

Nesta forma de importação, estamos ligando o nome `nome-alternativo` ao módulo em questão. A partir de então, utilizamos este nome alternativo para nos referir ao módulo.

```python
rand
```

```python
import random as rand
```

```python
rand

dir(rand)
```

```python
rand.choice([1,2,3,4])
```
