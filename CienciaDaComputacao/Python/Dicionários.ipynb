# Dicionários

## Características

- Estruturas de mapeamento
- Mutáveis
- Aceitam diferentes tipos de dados
- Não ordenadas

## Construtores

- `dict()`
- `{}`

```python
dict(Fulano=111, Beltrano=222, Ciclano=333)
```

```python
{"Fulano":111, "Beltrano":222}
```

```python
# Conjuntos:
{el1, el2, ...}

# Dicionários:
{k:v, k2:v2, ...}
```

## Acessando e Modificando

```python
d = {"Nome":"Augusto", "Idade":23, "País":"Brasil", "Habilitação":False}

print(d)
```

```python
print(d["Nome"])
print(d["Idade"])
print(d["País"])
print(d["Habilitação"])
```

```python
d["Idade"] = 35

print(d)
```

## Métodos

### Keys

Retorna todas as key dos dicionários.

```python
print(d.keys())

print(type(d.keys()))

print(list(d.keys()))
```

### Values

Retorna todos os valores do dicionário.

```python
print(d.values())
print(type(d.values()))

print(list(d.values()))
```

### Items

Retorna todos os pares key-valor.

```python
print(d.items())
print(type(d.items()))

print(list(d.items()))
```

### Fromkeys

Retorna um dicionário a partir das keys e valor especificado.

```python
dict.fromkeys(["Nome", "Idade", "País", "Habilitação"], False)
```
