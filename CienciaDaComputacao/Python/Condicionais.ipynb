# Estruturas Condicionais

Tais estruturas alteram o fluxo do programa, ou seja, alteram quais códigos são executados e quando são executados.

## If

Todos os comandos dentro desta estrutura serão executados somente quando a expressão for `True`.

```python
# Sintaxe:

if condição:
    código
```

```python
var = 2

if var < 3:
    print(f"O valor da variável var é {var}")
```

## Else

Os comandos dentro desta estrutura serão executados somente quando todas as condições anteriores forem `False`.

```python
# Sintaxe:

if condição:
    código
else:
    código
```

```python
var = 2

if var > 3:
    print(var)
else:
    print("O valor da variável var é menor que três!")
```

## Elif

`If`'s alternativos. Os comandos dentro desta estrutura serão executados somente quando as condições anteriores forem `False` e sua condição for `True`.

```python
# Sintaxe:

if condição:
    código
elif condição2:
    código
elif condição3:
    código
else:
    código
```

```python
var = 0

if var > 0:
    print("Var > 0")
elif var < 0:
    print("Var < 0")
else:
    print("Var == 0")

print("Depois da estrutura condicional.")
```

## Estruturas Independentes

```python
if 5 > 2:
    print(False)

if 5 > 3:
    print(True)
```
