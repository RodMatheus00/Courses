# Variáveis

- Variáveis são nomes que estão ligados a objetos na memória e fazem referência a estes.
- Operador = : Designação de variáveis
- Operador +=: iadd (adição-in-place)

```python
# Variável inteira:
numeros_pessoas = 16

print(numeros_pessoas)
```

```python
# Variável string:
linguagem = "Python"

print(linguagem)
print(type(linguagem))
```

```python
print(int(numeros_pessoas / 2))
```

```python
numeros_pessoas += 4

print(numeros_pessoas)
```

```python
numeros_pessoas *= 10

print(numeros_pessoas)
```

```python
numeros_pessoas -= 200

print(numeros_pessoas)
```

```python
print(5 + 5)

# F-string:

print(f"Resultado de 5 + 5 = {5 + 5}")

print(f"Variável linguagem: {linguagem}")
```
